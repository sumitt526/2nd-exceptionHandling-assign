Qsn 3 solution:-

ArithmeticError class is the base for all the errors associated with the airthmetic operations in python.
it has several subclasses such as ZeroDivisionError, FloatingPointError, OverFlowError, etc.

For example:-

ZeroDivisionError:-> it basically occurs when we divide any number with zero.

OverflowError:-> it occurs when a calculation produces a result that is too large to be represented

practical exmaple are listed below:->

```python
#for ZeroDivisionError:-

try:
    10/0
except ZeroDivisionError as e:
    print(e, e.__class__)
```

```python
#for OverflowError:-

try:
    z = 8.3
    for i in range(1, 1000):
        z = z**i
except ArithmeticError as e:
    print(e, e.__class__)
        
```

Qsn 4 solution:-

LookupError class is used when something can't be found or is out of range and it has a subclasses like indexerror and key error.
Now, let's explain both of these errors.

1) IndexError:-> whenever we are trying to access string or list out of range it will give IndexError.

2) KeyError:-> whenever we are trying to access the data by specifying the key and the key is not available in the dictionary then we will get keyerror.

practical examples are given below for each explained error:->



```python
#for IndexError :-

try:
    l = [10,2,4,33,24]
    l[5]
except Exception as e:
    print(e, e.__class__)
```

```python
#for KeyError:-

try:
    a = {"hare":"krishna", "hare":"rama", "recite": 10}
    print(a["radhe"])
except Exception as e:
    print(e, e.__class__)
```

ImportError :-> these are the errors which occurs when the module to be imported is not available or was not created moreover, it may also occur when import class is in circular dependency or may be misspelled.

ModuleNotFoundError:-> it is the syntax error that basically occurs when the import statement find the file at the declared path. 

```python
#Example;

try:
    import funnysingh
except ModuleNotFoundError as e:
    print(e, e.__class__)
```
